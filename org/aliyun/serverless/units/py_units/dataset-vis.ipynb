```python
import math

import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import os

path = "/Users/yeyuan/Desktop/dataSet_1/roles1_assign_25015.txt"

list = []
with open(path, "r") as f:
    for line in f:
        list.append(int(line.split(",")[1]))

data = pd.DataFrame({"x": range(len(list)), "y": list})
sns.lineplot(x="x", y="y", data=data)
plt.show()
```

```python
import math

log_list = [math.log1p(i) for i in list]
data = pd.DataFrame({"x": range(len(log_list)), "y": log_list})
sns.lineplot(x="x", y="y", data=data)
plt.show()
```
